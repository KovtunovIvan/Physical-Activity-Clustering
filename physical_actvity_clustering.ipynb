```python
#построим корреляционную матрицу
from pandas.plotting import scatter_matrix
import numpy as np

df_corr = df.corr()

f,ax = plt.subplots(figsize=(15,10))
sns.heatmap(df_corr, mask=np.zeros_like(df_corr, dtype = np.bool), cmap = 'BrBG', ax=ax)
plt.show()
```

```python
corr_features=[]
for i in range(len(df_corr.columns)):
    for j in range(i):
        if abs(df_corr.iloc[i,j])>0.8:
            colname=df_corr.columns[i]
            corr_features.append(colname)
```

```python
#произведем предобработку данных
df = df.drop(columns=['timestamp', 'subject_id'])
```

```python
#в качестве варианта заполения пропусков можно использовать KNNImputer
from sklearn.impute import KNNImputer

imputer = KNNImputer(n_neighbors=25)
imputer.fit_transform(df)

df.isna().sum()
```

```python
#используем в качестве заполнителя пропусков интерполяцию
df = df.interpolate()
df.isna().sum()
```

```python
#масштабируем данные
from sklearn.preprocessing import StandardScaler

df_std = StandardScaler().fit_transform(df)

scaled_df = pd.DataFrame(df_std, index=df.index, columns=df.columns)
scaled_df
```

```python
#понизим размерность признакового пространства
from sklearn.decomposition import PCA

pca = PCA(n_components=15)

principalComponents = pca.fit_transform(scaled_df)
principalDf = pd.DataFrame(data = principalComponents)
```

```python
principalDf
```

```python
#построим матрицу корреляции для обработанных данных
principalDf_corr = principalDf.corr()

f,ax = plt.subplots(figsize=(15,10))
sns.heatmap(principalDf_corr, mask=np.zeros_like(principalDf_corr, dtype = np.bool), cmap = 'YlGnBu', ax=ax)
plt.show()
```

```python
#посмотрим как сработала PCA
print(pca.explained_variance_ratio_)
evr = pca.explained_variance_ratio_
evr.sum()
```

```python
#повторно используем PCA, но уже будем обучать по проценту сохраненной информации
from sklearn.decomposition import PCA
pca = PCA(.85)
principalComponents = pca.fit_transform(scaled_df)
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA().fit(scaled_df)

plt.figure(figsize=(10,7))
plt.plot(np.cumsum(pca.explained_variance_ratio_), color='k', lw=2)
plt.xlabel('Number of components')
plt.ylabel('Total explained variance')
plt.xlim(0, 63)
plt.yticks(np.arange(0, 1.1, 0.1))
plt.axvline(24, c='b')
plt.axhline(0.9, c='r')
plt.show();
```

```python
pca = PCA(n_components=25)

principalComponents = pca.fit_transform(scaled_df)
principalDf = pd.DataFrame(data = principalComponents)
print(pca.explained_variance_ratio_)
evr = pca.explained_variance_ratio_
evr.sum()
```

```python
#кластеризация
from sklearn.cluster import KMeans

n_init = 12
max_iter = 225
tol = 0.0001
random_state = 123
n_clusters = 6

clusterer = KMeans(n_clusters=n_clusters, max_iter = max_iter, random_state = random_state)

cluster_labels = clusterer.fit_predict(principalComponents)

# inertia
inertia = clusterer.inertia_
inertia
```
